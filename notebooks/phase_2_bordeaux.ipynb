## Outliers

```python
def drop_outliers(df:pd.DataFrame, column:str):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    IQR=q3-q1

    without_outliers = df[((df[column]>(q1-1.5*IQR)) & (df[column]<(q3+1.5*IQR)))]
    return without_outliers
    
```

## 6. **Comparer des performances entre Lille et Bordeaux pour chaque type de logement** :
    - Le modèle n'est pas si performant sur Bordeaux.
    - Les écarts de performance 
      MSE XGBRegressor
                  Appart |Maison  
      Lille       | 0.21  | 0.85
      Bordeaux    | 0.50  | 11.00
    
    - Les facteurs qui peuvent expliquer ces différences: plusieurs facteurs structurels, statistiques ou liés aux données, d'une manière générale

      - plusieurs facteurs structurels, statistiques ou liés aux données, d'une manière générale
      -->Différences de distribution des données: Le modèle a appris sur un type de données, mais l’autre ville suit un comportement statistiquement différent.
      -->Spécificité géographique (features locales): Les variables utilisées sont trop spécifiques à la ville d’origine.
      -->Modèle trop complexe / surapprentissage: Le modèle a mémorisé les détails de Lille, sans apprendre des règles globales-->  Solution :Réduction de max_depth, Ajout de reg_lambda, subsample, early_stopping, Cross-validation rigoureuse
      -->Features manquantes ou mal encodées: Certaines variables peuvent avoir un sens différent d’une ville à l’autre ou être manquantes (ex:vue sur mer)--> Solution : Harmoniser les features, les encodages, vérifier les valeurs manquantes
      -->Échelle et normalisation : Si tu scales les données (MinMax, StandardScaler), il faut appliquer le même scaler appris sur la ville d'entraînement.
      -->Échantillon trop petit ou non représentatif: Le modèle a vu trop peu de cas variés à l’entraînement
      
                Appart |Maison
      Lille     |(49, 4) |  (326, 4)
      Bordeaux  |(56, 4) |  (362, 4)
      
    - Le modèle généralise-t-il mieux sur un type de bien que sur l'autre ?
      Non, les performances sont dégradées pour les appart que pour les maisons

```python
# Chargement des modèles
models = joblib.load('../models/models_appart_lille.pkl')
models_m = joblib.load('../models/models_maison_lille.pkl')

mse_results_appart = {}
mse_results_maison = {}

# Itération simultanée sur les deux dictionnaires
for name in models.keys():
    model_appart = models[name]
    model_maison = models_m[name]

    print(f"{name} (appart): {type(model_appart)}")
    print(f"{name} (maison): {type(model_maison)}")

    # Prédictions
    preds_a = model_appart.predict(Xa_scaled)
    preds_m = model_maison.predict(Xm_scaled)

    # Calcul MSE
    mse_a = mean_squared_error(ya_scaled, preds_a)
    mse_m = mean_squared_error(ym_scaled, preds_m)

    # Stockage résultats
    mse_results_appart[name] = mse_a
    mse_results_maison[name] = mse_m
```

```python
import matplotlib.pyplot as plt

# Données
labels_appart = list(mse_results_appart.keys())
mse_list_appart = list(mse_results_appart.values())

labels_maison = list(mse_results_maison.keys())
mse_list_maison = list(mse_results_maison.values())

fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Titre général
fig.suptitle("📊 Performance des modèles sur les données de test - Ville de Bordeaux", fontsize=16, fontweight='bold')

# Subplot 1 – Appartements
x_appart = range(len(labels_appart))
bars_appart = axes[0].bar(x_appart, mse_list_appart, color='skyblue')
axes[0].set_title("MSE par modèle (Appartements)")
axes[0].set_ylabel("MSE")
axes[0].set_xlabel("Modèles")
axes[0].set_xticks(x_appart)
axes[0].set_xticklabels(labels_appart, rotation=45, ha='right')
axes[0].grid(axis='y', linestyle='--', alpha=0.5)

for bar in bars_appart:
    height = bar.get_height()
    axes[0].text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:.2f}',
        ha='center',
        va='bottom'
    )

# Subplot 2 – Maisons
x_maison = range(len(labels_maison))
bars_maison = axes[1].bar(x_maison, mse_list_maison, color='salmon')
axes[1].set_title("MSE par modèle (Maisons)")
axes[1].set_xlabel("Modèles")
axes[1].set_xticks(x_maison)
axes[1].set_xticklabels(labels_maison, rotation=45, ha='right')
axes[1].grid(axis='y', linestyle='--', alpha=0.5)

for bar in bars_maison:
    height = bar.get_height()
    axes[1].text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:.2f}',
        ha='center',
        va='bottom'
    )

# Ajuster l'espace pour laisser la place au titre général
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python

import json

# Charger les MSE de Lille depuis le fichier JSON
with open("mse_lille.json", "r") as f:
    mse_lille = json.load(f)

# Données Bordeaux (déjà chargées dans mse_results_appart et mse_results_maison)
labels_appart = list(mse_results_appart.keys())
mse_list_bx_appart = list(mse_results_appart.values())
mse_list_lille_appart = [mse_lille["appartements"][label] for label in labels_appart]

labels_maison = list(mse_results_maison.keys())
mse_list_bx_maison = list(mse_results_maison.values())
mse_list_lille_maison = [mse_lille["maisons"][label] for label in labels_maison]

# ========================
# FIGURE : Appartements
# ========================
x = range(len(labels_appart))
bar_width = 0.35

fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
fig.suptitle("📊 Performance des modèles : Bordeaux vs Lille", fontsize=16, fontweight='bold')

# Subplot 1 : Appartements
axes[0].bar([i - bar_width/2 for i in x], mse_list_bx_appart, width=bar_width, label='Bordeaux', color='skyblue')
axes[0].bar([i + bar_width/2 for i in x], mse_list_lille_appart, width=bar_width, label='Lille', color='orange')
axes[0].set_title("Appartements")
axes[0].set_ylabel("MSE")
axes[0].set_xlabel("Modèles")
axes[0].set_xticks(x)
axes[0].set_xticklabels(labels_appart, rotation=45, ha='right')
axes[0].grid(axis='y', linestyle='--', alpha=0.5)

# Subplot 2 : Maisons
x = range(len(labels_maison))
axes[1].bar([i - bar_width/2 for i in x], mse_list_bx_maison, width=bar_width, label='Bordeaux', color='salmon')
axes[1].bar([i + bar_width/2 for i in x], mse_list_lille_maison, width=bar_width, label='Lille', color='green')
axes[1].set_title("Maisons")
axes[1].set_xlabel("Modèles")
axes[1].set_xticks(x)
axes[1].set_xticklabels(labels_maison, rotation=45, ha='right')
axes[1].grid(axis='y', linestyle='--', alpha=0.5)

# Légende et mise en page
axes[0].legend()
axes[1].legend()
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Calcul du meilleur modèle pour appartements
best_model_appart = min(mse_results_appart, key=mse_results_appart.get)
best_mse_appart = mse_results_appart[best_model_appart]

# Calcul du meilleur modèle pour maisons
best_model_maison = min(mse_results_maison, key=mse_results_maison.get)
best_mse_maison = mse_results_maison[best_model_maison]

print(f"🏡 Meilleur modèle pour appartements : {best_model_appart} avec un MSE de {best_mse_appart:.4f}")
print(f"🏠 Meilleur modèle pour maisons : {best_model_maison} avec un MSE de {best_mse_maison:.4f}")
```
